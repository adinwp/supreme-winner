# **Pearson Correlation**

```python
sheet_url = 'https://docs.google.com/spreadsheets/d/1_48L3MNGud5BvXItKQesd5TnakLwRp0jp_7_V3miLPI/edit#gid=0'
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df = pd.read_csv(sheet_url_trf)
df
```

```python
df.info()
```

```python
df = df.iloc[:12]
df
```

```python
df['data_1'] = df['data_1'].astype(int)
df['data_2'] = df['data_2'].astype(int)
```

```python
fig, ax = plt.subplots()
ax.plot(df['data_1'], df['data_2'], linewidth=0, marker='s')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
r,p_value = stats.pearsonr(df['data_1'], df['data_2'])
print(r)
print(p_value)
```

```python
df.corr()
```

# **Spearman Correlation**

difference spearman vs pearson correlation https://towardsdatascience.com/clearly-explained-pearson-v-s-spearman-correlation-coefficient-ada2f473b8

```python
sheet_url = 'https://docs.google.com/spreadsheets/d/14Tb2yAPweuUU4LhcP1q8dfZOz6V1jumLfTJRVphcWCs/edit#gid=0'
sheet_url_trf = sheet_url.replace('/edit#gid=', '/export?format=csv&gid=')
df = pd.read_csv(sheet_url_trf)
df
```

```python
df.info()
```

```python
r,p_value = stats.spearmanr(df['Rank Math'], df['Rank Science'])
print(r)
print(p_value)
```

```python
df.corr()
```

```python
import seaborn as sns
corr = df.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    annot=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
